1. Consider a ground sensor at 35°N latitude & 108°W longitude. Let us define the sensor’s “field of regard” as the entire area that the sensor can “see” above the horizon from its location on the Earth. This can be visualized by drawing a plane tangent to the Earth at the point where the sensor is located (see illustration below).

   Now consider an object in Earth orbit. At a particular instant in time, we know that the object’s position vector from Earth’s center expressed in ECI coordinates is (5294.35, 3707.14, 2352.42) km and that θg (Greenwich sidereal time) is 5°.

   Is the object within the sensor’s field of regard at this instant?

### Solution:

To find if the object is within the sensors field of regard, that means we must find the "Up" coordinate in the TH frame. Now listing our variables, we have that

Let the Earth's radius be:

$$
R_e = 6378 \ \text{km}
$$

Greenwich sidereal angle:

$$
\theta_g = 5^\circ
$$

Sensor's latitude:

$$
L = 35^\circ \text{N}
$$

Sensor's longitude:

$$
\gamma = 108^\circ \text{W}
$$

Objects's position vector in ECI frame:

$$
\mathbf{r}_{o,\text{ECI}} = [5294.35,\ 3707.14,\ 2352.42] \ \text{km}
$$

Sensor's position vector in the Topocentric-Horizon (TH) frame:

$$
\mathbf{r}_{s,\text{TH}} = [0,\ 0,\ R_e]
$$

Local sidereal angle:

$$
\theta = \gamma + \theta_g
$$

Coordinate transformation matrix from TH frame to ECI frame:

$$
{Q_{\text{ECI}}^\text{TH}} =
\begin{bmatrix}
\sin(L) \cos(\theta) & \sin(L) \sin(\theta) & -\cos(L) \\
-\sin(\theta) & \cos(\theta) & 0 \\
\cos(L) \cos(\theta) & \cos(L) \sin(\theta) & \sin(L)
\end{bmatrix}
$$

Now using that, we can get the sensors location in ECI

$$
\mathbf{r}_{s,\text{ECI}} = {{Q_{\text{ECI}}^\text{TH}}}^T \mathbf{r}_{s,\text{TH}}
$$

$$
=
\begin{bmatrix}
-0.12902662 & 0.97437006 & -0.18426912 \\
-0.55887571 & -0.22495105 & -0.79815723 \\
-0.81915204 & 0 & 0.57357644
\end{bmatrix}^T
\begin{bmatrix}
0 \\
0 \\
6378
\end{bmatrix}
$$
$$
= \begin{bmatrix}
-1175.26842204 \\
-5090.64681589 \\
3658.27051105
\end{bmatrix} \text{ km}
$$

Now the vector from the sensor to the object is given as

$$
\bm{r}_{so, ECI} = \bm{r}_{o,ECI} - \bm{r}_{s,ECI}
$$
$$
= \begin{bmatrix}
5294.35 \\
3707.14 \\
2352.42
\end{bmatrix}
-
\begin{bmatrix}
-1175.26842204 \\
-5090.64681589 \\
3658.27051105
\end{bmatrix}
$$
$$
= 
\begin{bmatrix}
6469.61842204 \\
8797.78681589 \\
-1305.85051105
\end{bmatrix} \text{ km}
$$

Now to find the TH coordinate system of the object we convert this by

$$
\mathbf{r}_{s,\text{TH}} = {{Q_{\text{ECI}}^\text{TH}}}\mathbf{r}_{s,\text{ECI}}
$$

This gives us 

$$
\mathbf{r}_{so,\text{TH}} 
= 
\begin{bmatrix}
-4681.93225687 \\
4324.7311009 \\
-8963.17310939
\end{bmatrix} \text{ km}
$$

Now since the z-component is negative, that means that it's below the horizon so the object is not viewable.

2. Given the same sensor above at the same instant in time, consider a star at RA = 17°, Dec = 62°. Can
the sensor “see” the star at this instant?

### Solution:

Given Right Ascension and Declination, the unit vector in the ECI frame of the star is:

$$
\vec{\bm{r}}_{star, ECI} = 
\begin{bmatrix}
\cos(\delta) \cos(\alpha) \\
\cos(\delta) \sin(\alpha) \\
\sin(\delta)
\end{bmatrix} = 
\begin{bmatrix}
0.44895789\\
0.1372602\\
0.88294759
\end{bmatrix}
$$

The unit vector of the sensor is then simply

$$
\vec{\bm{r}}_{star, TH} = 
\begin{bmatrix}
0 \\
0 \\
1
\end{bmatrix}
$$

The rest of the steps are the same as before. We're using the same DCM since it's the same time. The only difference is we are now working with unit vectors. This gives us the unit vector pointing from the sensor to the star as 

$$
\vec{\bm{r}}_{ss, TH} = 
\begin{bmatrix}
-0.85790724 \\
0.4065743 \\
-0.68584636
\end{bmatrix}
$$

Since the z-component is negative, then it is not in the field of regard of the sensor.


```python
import numpy as np

# Problem 1

R_e = 6378
theta_g = np.radians(5)

lat = np.radians(35)
lon = np.radians(-108)

r_o_ECI = np.array([5294.35, 3707.14, 2352.42])
r_s_TH  = np.array([0, 0, R_e])

theta = lon + theta_g

Q_ECI_TH = np.array([[np.sin(lat)*np.cos(theta), np.sin(lat)*np.sin(theta), -np.cos(lat)],
                      [-np.sin(theta), np.cos(theta), 0],
                      [np.cos(lat)*np.cos(theta), np.cos(lat)*np.sin(theta), np.sin(lat)]])


r_s_ECI = np.transpose(Q_ECI_TH) @ r_s_TH
r_so_ECI = r_o_ECI - r_s_ECI
r_so_TH = Q_ECI_TH @ r_so_ECI

print(r_so_TH)

# Problem 2

RA = np.radians(17)
Dec = np.radians(62)

# Convert to ECI coordinates
x = np.cos(Dec) * np.cos(RA)
y = np.cos(Dec) * np.sin(RA)
z = np.sin(Dec)

r_o_ECI = np.array([x, y, z])
r_s_TH  = np.array([0, 0, 1])

r_s_ECI = np.transpose(Q_ECI_TH) @ r_s_TH
r_so_ECI = r_o_ECI - r_s_ECI
r_so_TH = Q_ECI_TH @ r_so_ECI

print(r_so_TH)
```
