**Question 1:**  
The primary quantity measured by a radar sensor is _______ .

**Group of answer choices:**  
A. Range  
B. Line of sight  
C. Range rate  
D. Angle rate  

**Answer:**  
**A. Range**

---

**Question 2:**  
The center of a camera's image plane is called the ________ .

**Group of answer choices:**  
A. Field of view  
B. Focal length  
C. Boresight  

**Answer:**  
**C. Boresight**

---

**Question 3:**  
The number of angles required to indicate the direction from a sensor to a space object is _____ .

**Group of answer choices:**  
A. 1  
B. 2  
C. 3  

**Answer:**  
**B. 2**

---

**Question 4:**  
The value of “bit depth” is always a power of ____ .

**Group of answer choices:**  
A. 2  
B. 3  
C. 4  

**Answer:**  
**A. 2**

---
