**Question 1:**  
The coordinate frame often used to chart the position of stars is the _______ Frame.

**Group of answer choices:**  
A. Geocentric Equatorial  
B. Topocentric Horizon  
C. Earth-centered Earth-fixed  
D. Celestial  

**Answer:**  
**D. Celestial**

---

**Question 2:**  
Which of the following coordinate frames does NOT have its origin at Earth's center?

**Group of answer choices:**  
A. Celestial  
B. Geocentric Equatorial  
C. Earth-centered Earth-fixed frame  
D. Topocentric Horizon Frame  

**Answer:**  
**D. Topocentric Horizon Frame**

---

**Question 3:**  
The vernal equinox direction is fixed with respect to the stars.

**Group of answer choices:**  
True  
False

**False**
