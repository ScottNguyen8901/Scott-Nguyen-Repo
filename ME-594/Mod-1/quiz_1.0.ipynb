**Question 1:**  
Newton’s Law of Gravitation involves the derivation of an equation for the gravitational attraction between/among ___ bodies.

**Group of answer choices:**  
A. 2  
B. 3  
C. 4  
D. all bodies in the universe  

**Answer:**  
**A. 2**

---

**Question 2:**  
Consider a small object orbiting about a larger body. Assuming the only force present is the gravitational attraction between the object and the body, the object’s angular momentum vector is ____ to its orbital plane.

**Group of answer choices:**  
A. parallel  
B. perpendicular  
C. neither  

**Answer:**  
**B. perpendicular**

---

**Question 3:**  
The time derivative of a unit vector taken in the unit vector’s own frame is ____.

**Group of answer choices:**  
A. 0  
B. 1  
C. it varies  

**Answer:**  
**A. 0**

---

**Question 4:**  
The time derivative of any vector as seen in frame “F1” is different from the time derivative as seen in frame “F2”, and this difference is based on the ____.

**Group of answer choices:**  
A. instantaneous orientation of F2 relative to F1  
B. instantaneous rotational velocity of F2 relative to F1  
C. angle between the x axis of F2 and the x axis of F1  

**Answer:**  
**B. instantaneous rotational velocity of F2 relative to F1**

---

**Question 5:**  
Based on the definition of the standard Earth-centered inertial frame, the “z” axis of this frame is aligned with ____.

**Group of answer choices:**  
A. Earth’s equatorial plane  
B. Earth’s polar axis  
C. Earth’s ecliptic plane  

**Answer:**  
**B. Earth’s polar axis**

---

**Question 6:**  
Consider 2 masses, m1 and m2, under the influence of each other's gravitational attraction. If m1 >> m2, the motion of m2 with respect to the center of mass of the system is ____.

**Group of answer choices:**  
A. the same as the motion of m1 with respect to the center of mass  
B. the same as the motion of m2 with respect to m1  
C. m2 doesn’t move with respect to the center of mass  

**Answer:**  
**B. the same as the motion of m2 with respect to m1**
