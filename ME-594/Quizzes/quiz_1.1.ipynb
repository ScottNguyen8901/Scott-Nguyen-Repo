**Question 1:**  
Other than Earth's gravity, the dominant force on an object in Low Earth orbit (1,000km altitude or below) is:

**Group of answer choices:**  
A. drag  
B. solar radiation pressure  
C. 3rd-body gravity (Sun, Moon)  
D. none of the above  

**Answer:**  
**A. drag**

---

**Question 2:**  
Of the forces listed below, the LEAST dominant force on an object in Geostationary Earth orbit (approx. 36,000km altitude) is:

**Group of answer choices:**  
A. drag  
B. solar radiation pressure  
C. 3rd body gravity (Sun, Moon)  

**Answer:**  
**A. drag**
